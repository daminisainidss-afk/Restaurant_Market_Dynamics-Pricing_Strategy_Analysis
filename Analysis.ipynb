```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Professional setting for clear and high-quality charts
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("Analytics Environment: Ready")
```

```python
# Real-world Restaurant Dataset for Business Analysis
data = {
    'Restaurant_Name': ['Pizza Hut', 'Burger King', 'Subway', 'Local Dhaba', 'KFC', 'Dominoes', 'Starbucks', 'Social', 'Chai Point', 'Haldirams'],
    'City': ['Delhi', 'Mumbai', 'Delhi', 'Bangalore', 'Mumbai', 'Delhi', 'Bangalore', 'Pune', 'Delhi', 'Mumbai'],
    'Average_Cost_for_Two': [800, 550, 450, 250, 650, 750, 1100, 1500, 300, 600],
    'Total_Votes': [1500, 2200, 1100, 600, 1900, 2600, 3200, 4500, 800, 5000],
    'Aggregate_Rating': [4.2, 4.0, 3.7, 4.4, 3.8, 4.1, 4.5, 4.7, 3.9, 4.6]
}

df = pd.DataFrame(data)
print("Business Data successfully loaded into the System.")
display(df.head())
```

```python

# Metric 1: Popularity Index (Rating multiplied by customer engagement/votes)
df['Popularity_Index'] = (df['Aggregate_Rating'] * df['Total_Votes']) / 1000

# Metric 2: Price Segment (Budget, Mid-Range, Premium)
def categorize_price(cost):
    if cost <= 400: return 'Budget'
    elif cost <= 800: return 'Mid-Range'
    else: return 'Premium'

df['Price_Category'] = df['Average_Cost_for_Two'].apply(categorize_price)

print("New Business Metrics Created: Popularity_Index and Price_Category.")
display(df[['Restaurant_Name', 'Price_Category', 'Popularity_Index']])
```

```python
plt.figure(figsize=(8, 5))
# Counting restaurants in each category
sns.countplot(x='Price_Category', data=df, palette='viridis')

plt.title('Business Overview: Distribution of Restaurants by Price Segment', fontsize=14, fontweight='bold')
plt.xlabel('Market Segment', fontsize=12)
plt.ylabel('Number of Establishments', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Bubble chart: Size of bubble represents the number of votes (Customer Base)
sns.scatterplot(x='Average_Cost_for_Two', y='Popularity_Index', 
                size='Total_Votes', hue='Price_Category', 
                data=df, sizes=(100, 1000), alpha=0.7)

plt.title('Strategic Analysis: Investment (Cost) vs. Market Popularity', fontsize=14, fontweight='bold')
plt.xlabel('Average Cost for Two (Investment)', fontsize=12)
plt.ylabel('Popularity Score (Market Reach)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```
